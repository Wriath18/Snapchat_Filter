```python

face_cascade = cv2.CascadeClassifier("frontalEyes35x16.xml")
mustache_cascade = cv2.CascadeClassifier("Nose18x15.xml")
sunglass_img = cv2.imread('glasses.png', -1)
mustache_image = cv2.imread("mustache.png", -1)

cam = cv2.VideoCapture(0)

while True:
    ret_value, frame = cam.read()
    if not ret_value:
        print("Error: Cannot read frame from the camera")
        break

    faces = face_cascade.detectMultiScale(frame, 1.3, 5)
    noses = mustache_cascade.detectMultiScale(frame, 1.3, 5)
    print("Number of faces detected:", len(faces))
    
    # for (x, y, w, h) in faces:
    #     (x2,y2,w2,h2) = [nose for nose in noses]
    #     cv2.rectangle(frame, (x2,y2),(x2+w2, y2+h2), 2)
    #     scaling_factor = 1.5
    #     resized_sunglass = cv2.resize(sunglass_img, (int(w*1.5), int(h*1.5)))
    #     print("Resized sunglasses image shape:", resized_sunglass.shape)
    #     for i in range(resized_sunglass.shape[0]):
    #         for j in range(resized_sunglass.shape[1]):
    #             if resized_sunglass[i, j, 3] != 0:
    #                 frame[y + i, x  + j - 25, :] = resized_sunglass[i, j, :-1]
    for (x, y, w, h) in faces:
        for (x2, y2, w2, h2) in noses:
            scaling_factor = 1.2
            resized_mustache = cv2.resize(mustache_image, (int(w2*scaling_factor), int(h2*scaling_factor)))
            resized_sunglass = cv2.resize(sunglass_img, (int(w * scaling_factor), int(h * scaling_factor)))
            print("Resized sunglasses image shape:", resized_sunglass.shape)
            for i in range(resized_mustache.shape[0]):
                for j in range(resized_mustache.shape[1]):
                    if resized_mustache[i, j, 3] != 0:
                        frame[y2 + i + int(0.6 * h2), x2 + j, :] = resized_mustache[i, j, :-1]
            for i in range(resized_sunglass.shape[0]):
                for j in range(resized_sunglass.shape[1]):
                    if resized_sunglass[i, j, 3] != 0:
                        frame[y + i, x + j - 10, :] = resized_sunglass[i, j, :-1]
                        
    

    cv2.imshow("Video Frame", frame)

    key_pressed = cv2.waitKey(1) & 0xFF
    if key_pressed == ord("s"):
        break

cam.release()
cv2.destroyAllWindows()
```

```python
cam = cv2.VideoCapture(0)
face_cascade = cv2.CascadeClassifier("haarcascade_frontalface_alt.xml")


while True:
    ret_value, frame = cam.read()
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    if ret_value == False:
        continue
    faces = mustache_cascade.detectMultiScale(gray_frame,1.3,5)


    for (x,y,w,h) in faces:
        cv2.rectangle(frame, (x,y), (x+w, y+h), (245,255,0),2)

    cv2.imshow("Video Frame", frame)

    key_pressed = cv2.waitKey(1) & 0xFF
    if key_pressed == ord("s"):
        break

cam.release()
cv2.destroyAllWindows() 
```
