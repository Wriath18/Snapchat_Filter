```python

face_cascade = cv2.CascadeClassifier("frontalEyes35x16.xml")
sunglass_img = cv2.imread('glasses.png', -1)

cam = cv2.VideoCapture(0)

while True:
    ret_value, frame = cam.read()
    if not ret_value:
        print("Error: Cannot read frame from the camera")
        break

    faces = face_cascade.detectMultiScale(frame, 1.3, 5)
    print("Number of faces detected:", len(faces))

    for (x, y, w, h) in faces:
        scaling_factor = 1.2
        resized_sunglass = cv2.resize(sunglass_img, (int(w*scaling_factor), int(h*scaling_factor)))
        print("Resized sunglasses image shape:", resized_sunglass.shape)
        for i in range(resized_sunglass.shape[0]):
            for j in range(resized_sunglass.shape[1]):
                if resized_sunglass[i, j, 3] != 0:
                    frame[y + i, x  + j - 10, :] = resized_sunglass[i, j, :-1]

                        
    
    cv2.imshow("Video Frame", frame)

    key_pressed = cv2.waitKey(1) & 0xFF
    if key_pressed == ord("s"):
        break

cam.release()
cv2.destroyAllWindows()
```
